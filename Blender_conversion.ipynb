# **Convert Single .off file to .ply format**

```python
import trimesh


input_file = 'input_filename'
output_file = input_filename.replace('.off', '.ply')

# Load and export
mesh = trimesh.load(input_filename)
mesh.export(output_file)


print(f"‚úÖ Converted: {input_filename} ‚Üí {output_file}")
```

Convert Multiple Images

```python
import os
import trimesh

os.makedirs("converted_ply", exist_ok=True)

for file in os.listdir():
    if file.endswith(".off"):
        mesh = trimesh.load(file)
        out_path = os.path.join("converted_ply", file.replace(".off", ".ply"))
        mesh.export(out_path)
        print(f"‚úÖ Converted: {file} ‚Üí {out_path}")
```
